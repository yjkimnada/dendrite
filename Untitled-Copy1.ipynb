```python
import numpy as np
import torch 
import torch.nn as nn
from tqdm import tnrange
import torch.optim as optim
import torch.nn.functional as F
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.metrics import explained_variance_score
import scipy
from sklearn.linear_model import LinearRegression
```

```python
base_dir = "/media/hdd01/sklee/"
experiment = "clust4-60"
cell_type = "CA1"

E_neural_file = "Espikes_neural.npz"

V_file = "V_diff.npy"

E_neural = scipy.sparse.load_npz(base_dir+cell_type+"_"+experiment+"/data/"+E_neural_file)[:500000]
V = np.load(base_dir+cell_type+"_"+experiment+"/data/"+V_file)

E_neural = E_neural.toarray()
```

```python
eloc = np.load("/media/hdd01/sklee/CA1_clust4-60/data/Elocs_T10_Ne2000_gA0.6_tauA1_gN0.8_Ni200_gG0.1_gB0.1_Er0.5_Ir7.4_random_NR_rep150_stimseed1.npy")

den = np.unique(eloc[880:1120,0])
e_idx = np.where(np.isin(eloc[:,0], den) == True)[0]
print(e_idx.size)

C_syn_e = np.zeros((12,2000))

for i in range(12):
    C_syn_e[i,880+20*i:880+20*(i+1)] = 1
```

```python
S = np.matmul(E_neural, C_syn_e.T)
S_pad = np.zeros((500000+100-1, 12))
S_pad[99:] = S

t = np.tile(np.arange(100).reshape((1,-1)), (12,1))
t_tau = t / 5
kern = t_tau * np.exp(-t_tau)
kern = np.flip(kern, 1)

kern = torch.from_numpy(kern.copy()).unsqueeze(1)
S_pad = torch.from_numpy(S_pad).T.unsqueeze(0)

S_conv = F.conv1d(S_pad, kern, groups=12).squeeze(0).T.cpu().detach().numpy()
print(S_conv.shape)
```

```python
reg = LinearRegression().fit(S_conv[:500000], V[:500000].reshape(-1,1))
weights = reg.coef_.flatten()
print(reg.score(S_conv[:500000], V[:500000].reshape(-1,1)))
```
