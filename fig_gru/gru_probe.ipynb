```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
na = np.load("/media/hdd01/sklee/CA1_clust4-60/hand/gru_s10_h20_probe.npy")
nona = np.load("/media/hdd01/sklee/CA1_clust4-60_noNA/hand/gru_s10_h20_probe.npy")
```

```python
fig, axs = plt.subplots(nrows=2, ncols=6, figsize = (20,6))

for i in range(4):
    axs[0,0].plot(np.arange(0,100,0.2), nona[i,0,-1,500:1000] - np.min(nona[i,0,-1,500:1000]), label="Non-clust Subunit "+str(i+1))
    axs[0,1].plot(np.arange(1,31,1), np.max(nona[i,0,:,500:] - np.min(nona[i,0,-1,500:1000]), 1), label="Non-clust Subunit "+str(i+1))
    axs[1,0].plot(np.arange(0,100,0.2), nona[i,2,-1,500:1000] - np.min(nona[i,2,-1,500:1000]), label="Non-clust Subunit "+str(i+1))
    axs[1,1].plot(np.arange(1,31,1), np.max(nona[i,2,:,500:] - np.min(nona[i,2,-1,500:]), 1), label="Non-clust Subunit "+str(i+1))
            
    axs[0,2].plot(np.arange(0,100,0.2), nona[i+4,0,-1,500:1000] - np.min(nona[i+4,0,-1,500:1000]), label="Non-clust Subunit "+str(i+1))
    axs[0,3].plot(np.arange(1,31,1), np.max(nona[i+4,0,:,500:] - np.min(nona[i+4,0,-1,500:1000]), 1), label="Non-clust Subunit "+str(i+1))
    axs[1,2].plot(np.arange(0,100,0.2), nona[i+4,2,-1,500:1000] - np.min(nona[i+4,2,-1,500:1000]), label="Non-clust Subunit "+str(i+1))
    axs[1,3].plot(np.arange(1,31,1), np.max(nona[i+4,2,:,500:] - np.min(nona[i+4,2,-1,500:1000]), 1), label="Non-clust Subunit "+str(i+1))
            
    axs[0,4].plot(np.arange(0,100,0.2), na[i,0,-1,500:1000], label="Clust Subunit "+str(i+1))
    axs[0,5].plot(np.arange(1,31,1), np.max(na[i,0,:,500:], 1), label="Clust Subunit "+str(i+1))
    axs[1,4].plot(np.arange(0,100,0.2), na[i,2,-1,500:1000], label="Clust Subunit "+str(i+1))
    axs[1,5].plot(np.arange(1,31,1), np.max(na[i,2,:,500:], 1), label="Clust Subunit "+str(i+1))
    


axs[0,0].scatter(np.arange(0,0+2*31,2),np.ones(31)*0.1, marker="|", s=100, color="black")
axs[1,0].scatter(np.arange(0,0+30*31,30),np.ones(31)*0.025, marker="|", s=100, color="black")
axs[0,2].scatter(np.arange(0,0+2*31,2),np.ones(31)*0.1, marker="|", s=100, color="black")
axs[1,2].scatter(np.arange(0,0+30*31,30),np.ones(31)*0, marker="|", s=100, color="black")
axs[0,4].scatter(np.arange(0,0+2*31,2),np.ones(31)*(-0.25), marker="|", s=100, color="black")
axs[1,4].scatter(np.arange(0,0+30*31,30),np.ones(31)*(-0.0065), marker="|", s=100, color="black")

for i in range(3):
    axs[0,i*2].set_xlim(0,100)
    axs[1,i*2].set_xlim(0,100)
    

for i in range(3):
    axs[0,i*2].set_ylabel("Subunit output (a.u.)")
    axs[1,i*2].set_ylabel("Subunit output (a.u.)")
    axs[0,i*2+1].set_ylabel("Maximum output value (a.u.)")
    axs[1,i*2+1].set_ylabel("Maximum output value (a.u.)")
    axs[0,i*2].set_xlabel("Time (ms)")
    axs[0,i*2+1].set_xlabel("Input spike count")
    axs[1,i*2].set_xlabel("Time (ms)")
    axs[1,i*2+1].set_xlabel("input spike count")
    
    axs[0,i*2].legend()
    
axs[0,0].set_title("No-Na+ clustered subunits\n(2ms intervals)")
axs[1,0].set_title("No-Na+ clustered subunits\n(30ms intervals)")
axs[0,2].set_title("No-Na+ non-clustered subunits\n(2ms intervals)")
axs[1,2].set_title("No-Na+ non-clustered subunits\n(30ms intervals)")
axs[0,4].set_title("Na+-Diff clustered subunits\n(2ms intervals)")
axs[1,4].set_title("Na+-Diff clustered subunits\n(30ms intervals)")

fig.tight_layout()
#plt.show()
#fig.savefig("/home/sklee/dendrite/fig_gru/gru_probe.pdf", bbox_inches="tight", transparent=True)
```
