```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tnrange
from sklearn import metrics
from sklearn.metrics import precision_recall_curve
import matplotlib as mpl
```

```python
length = int(50000/2/5)
ap_true_asoma = np.load("/media/hdd01/sklee/CA1_clust4-60_aSoma/ap_true/gru_l20_h40_2ms_test.npy").flatten()
na_true_asoma = np.load("/media/hdd01/sklee/CA1_clust4-60_aSoma/na_true/gru_l20_h40_2ms_test.npy").flatten()
nona_true_asoma = np.load("/media/hdd01/sklee/CA1_clust4-60_aSoma/nona_true/gru_l20_h40_2ms_test.npy").flatten()
comb_pred_asoma = np.load("/media/hdd01/sklee/CA1_clust4-60_aSoma/comb_pred/gru_l20_h40_2ms_test.npy").flatten()
whole_pred_asoma = np.load("/media/hdd01/sklee/CA1_clust4-60_aSoma/whole_pred/gru_l20_h40_2ms_test.npy").flatten()
spk_asoma = np.load("/media/hdd01/sklee/CA1_clust4-60_aSoma/data/spike_train_2ms.npy")[980*length:].flatten()

ap_true_active = np.load("/media/hdd01/sklee/CA1_clust4-60_AP/ap_true/gru_l20_h40_2ms_test.npy").flatten()
na_true_active = np.load("/media/hdd01/sklee/CA1_clust4-60_AP/na_true/gru_l20_h40_2ms_test.npy").flatten()
nona_true_active = np.load("/media/hdd01/sklee/CA1_clust4-60_AP/nona_true/gru_l20_h40_2ms_test.npy").flatten()
comb_pred_active = np.load("/media/hdd01/sklee/CA1_clust4-60_AP/comb_pred/gru_l20_h40_2ms_test.npy").flatten()
whole_pred_active = np.load("/media/hdd01/sklee/CA1_clust4-60_AP/whole_pred/gru_l20_h40_2ms_test.npy").flatten()
spk_active = np.load("/media/hdd01/sklee/CA1_clust4-60_AP/data/spike_train_2ms.npy")[980*length:].flatten()
```

```python
prec_comb_pred_asoma, rec_comb_pred_asoma, _ = precision_recall_curve(spk_asoma, comb_pred_asoma)
prec_whole_pred_asoma, rec_whole_pred_asoma, _ = precision_recall_curve(spk_asoma, whole_pred_asoma)
prec_ap_true_asoma, rec_ap_true_asoma, _ = precision_recall_curve(spk_asoma, ap_true_asoma)
prec_na_true_asoma, rec_na_true_asoma, _ = precision_recall_curve(spk_asoma, na_true_asoma)
prec_nona_true_asoma, rec_nona_true_asoma, _ = precision_recall_curve(spk_asoma, nona_true_asoma)
    
prarea_comb_pred_asoma = metrics.auc(rec_comb_pred_asoma, prec_comb_pred_asoma)
prarea_whole_pred_asoma = metrics.auc(rec_whole_pred_asoma, prec_whole_pred_asoma)
prarea_ap_true_asoma = metrics.auc(rec_ap_true_asoma, prec_ap_true_asoma)
prarea_na_true_asoma = metrics.auc(rec_na_true_asoma, prec_na_true_asoma)
prarea_nona_true_asoma = metrics.auc(rec_nona_true_asoma, prec_nona_true_asoma)

prec_comb_pred_active, rec_comb_pred_active, _ = precision_recall_curve(spk_active, comb_pred_active)
prec_whole_pred_active, rec_whole_pred_active, _ = precision_recall_curve(spk_active, whole_pred_active)
prec_ap_true_active, rec_ap_true_active, _ = precision_recall_curve(spk_active, ap_true_active)
prec_na_true_active, rec_na_true_active, _ = precision_recall_curve(spk_active, na_true_active)
prec_nona_true_active, rec_nona_true_active, _ = precision_recall_curve(spk_active, nona_true_active)
    
prarea_comb_pred_active = metrics.auc(rec_comb_pred_active, prec_comb_pred_active)
prarea_whole_pred_active = metrics.auc(rec_whole_pred_active, prec_whole_pred_active)
prarea_ap_true_active = metrics.auc(rec_ap_true_active, prec_ap_true_active)
prarea_na_true_active = metrics.auc(rec_na_true_active, prec_na_true_active)
prarea_nona_true_active = metrics.auc(rec_nona_true_active, prec_nona_true_active)

prarea_comb_pred_asoma_array = np.zeros((20))
prarea_whole_pred_asoma_array = np.zeros((20))
prarea_ap_true_asoma_array = np.zeros((20))
prarea_na_true_asoma_array = np.zeros((20))
prarea_nona_true_asoma_array = np.zeros((20))

prarea_comb_pred_active_array = np.zeros((20))
prarea_whole_pred_active_array = np.zeros((20))
prarea_ap_true_active_array = np.zeros((20))
prarea_na_true_active_array = np.zeros((20))
prarea_nona_true_active_array = np.zeros((20))

for i in tnrange(20):
    part_prec_comb_pred_asoma, part_rec_comb_pred_asoma, _ = precision_recall_curve(spk_asoma[i*length:(i+1)*length], comb_pred_asoma[i*length:(i+1)*length])
    part_prec_whole_pred_asoma, part_rec_whole_pred_asoma, _ = precision_recall_curve(spk_asoma[i*length:(i+1)*length], whole_pred_asoma[i*length:(i+1)*length])
    part_prec_ap_true_asoma, part_rec_ap_true_asoma, _ = precision_recall_curve(spk_asoma[i*length:(i+1)*length], ap_true_asoma[i*length:(i+1)*length])
    part_prec_na_true_asoma, part_rec_na_true_asoma, _ = precision_recall_curve(spk_asoma[i*length:(i+1)*length], na_true_asoma[i*length:(i+1)*length])
    part_prec_nona_true_asoma, part_rec_nona_true_asoma, _ = precision_recall_curve(spk_asoma[i*length:(i+1)*length], nona_true_asoma[i*length:(i+1)*length])
    
    prarea_comb_pred_asoma_array[i] = metrics.auc(part_rec_comb_pred_asoma, part_prec_comb_pred_asoma)
    prarea_whole_pred_asoma_array[i] = metrics.auc(part_rec_whole_pred_asoma, part_prec_whole_pred_asoma)
    prarea_ap_true_asoma_array[i] = metrics.auc(part_rec_ap_true_asoma, part_prec_ap_true_asoma)
    prarea_na_true_asoma_array[i] = metrics.auc(part_rec_na_true_asoma, part_prec_na_true_asoma)
    prarea_nona_true_asoma_array[i] = metrics.auc(part_rec_nona_true_asoma, part_prec_nona_true_asoma)
    
    part_prec_comb_pred_active, part_rec_comb_pred_active, _ = precision_recall_curve(spk_active[i*length:(i+1)*length], comb_pred_active[i*length:(i+1)*length])
    part_prec_whole_pred_active, part_rec_whole_pred_active, _ = precision_recall_curve(spk_active[i*length:(i+1)*length], whole_pred_active[i*length:(i+1)*length])
    part_prec_ap_true_active, part_rec_ap_true_active, _ = precision_recall_curve(spk_active[i*length:(i+1)*length], ap_true_active[i*length:(i+1)*length])
    part_prec_na_true_active, part_rec_na_true_active, _ = precision_recall_curve(spk_active[i*length:(i+1)*length], na_true_active[i*length:(i+1)*length])
    part_prec_nona_true_active, part_rec_nona_true_active, _ = precision_recall_curve(spk_active[i*length:(i+1)*length], nona_true_active[i*length:(i+1)*length])
    
    prarea_comb_pred_active_array[i] = metrics.auc(part_rec_comb_pred_active, part_prec_comb_pred_active)
    prarea_whole_pred_active_array[i] = metrics.auc(part_rec_whole_pred_active, part_prec_whole_pred_active)
    prarea_ap_true_active_array[i] = metrics.auc(part_rec_ap_true_active, part_prec_ap_true_active)
    prarea_na_true_active_array[i] = metrics.auc(part_rec_na_true_active, part_prec_na_true_active)
    prarea_nona_true_active_array[i] = metrics.auc(part_rec_nona_true_active, part_prec_nona_true_active)
```

```python
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['font.sans-serif'] = "Arial"
mpl.rcParams['font.family'] = "sans-serif"

fig = plt.figure(constrained_layout=True, figsize = (12,12))
gs = fig.add_gridspec(4,2, wspace=0.01, hspace=0.05)
ax1 = fig.add_subplot(gs[0,:])
ax2 = fig.add_subplot(gs[1,0])
ax3 = fig.add_subplot(gs[1,1])
ax4 = fig.add_subplot(gs[2,:])
ax5 = fig.add_subplot(gs[3,0])
ax6 = fig.add_subplot(gs[3,1])

ax1.scatter(np.arange(0,500,2), spk_active[2350:2600]*2-0.8, marker="|", s=250, color="black", label="Measured APs")
ax1.plot(np.arange(0,500,2), ap_true_active[2350:2600], linewidth=1, color="black", alpha=1, label="AP, NA True")
ax1.plot(np.arange(0,500,2), na_true_active[2350:2600], linewidth=1, color="blue", alpha=0.75, label="NA True")
ax1.plot(np.arange(0,500,2), nona_true_active[2350:2600], linewidth=1, color="green", alpha=0.75, label="No NA True")
ax1.plot(np.arange(0,500,2), comb_pred_active[2350:2600], linewidth=1, color="red", alpha=0.75, label="Comb Pred")
ax1.plot(np.arange(0,500,2), whole_pred_active[2350:2600], linewidth=1, color="purple", alpha=0.75, label="Whole Pred")
ax1.legend()
ax1.set_ylim(0,1.5)
ax1.set_xlabel("Time (ms)")
ax1.set_ylabel("AP Probability")
ax1.set_yticks([0,0.2,0.4,0.6,0.8,1])
ax1.spines.right.set_visible(False)
ax1.spines.top.set_visible(False)
ax1.spines.left.set_bounds((0, 1))

bp1 = ax2.boxplot(prarea_ap_true_active_array, positions=[1], widths = 0.6, patch_artist=True)
bp2 = ax2.boxplot(prarea_na_true_active_array, positions=[2], widths = 0.6, patch_artist=True)
bp3 = ax2.boxplot(prarea_nona_true_active_array, positions=[3], widths = 0.6, patch_artist=True)
bp4 = ax2.boxplot(prarea_comb_pred_active_array, positions=[4], widths = 0.6, patch_artist=True)
bp5 = ax2.boxplot(prarea_whole_pred_active_array, positions=[5], widths = 0.6, patch_artist=True)
bp1['boxes'][0].set_facecolor("black")
bp2['boxes'][0].set_facecolor("blue")
bp3['boxes'][0].set_facecolor("green")
bp4['boxes'][0].set_facecolor("red")
bp5['boxes'][0].set_facecolor("purple")
bp1['boxes'][0].set_alpha(0.5)
bp2['boxes'][0].set_alpha(0.5)
bp3['boxes'][0].set_alpha(0.5)
bp4['boxes'][0].set_alpha(0.5)
bp5['boxes'][0].set_alpha(0.5)
bp1['medians'][0].set_color("black")
bp2['medians'][0].set_color("black")
bp3['medians'][0].set_color("black")
bp4['medians'][0].set_color("black")
bp5['medians'][0].set_color("black")
ax2.set_xticklabels(["AP, NA True","NA True", "No NA True","Comb Pred","Whole Pred"])
ax2.set_ylabel("Precision-Recall AUC")
ax2.spines.right.set_visible(False)
ax2.spines.top.set_visible(False)

ax3.plot(rec_ap_true_active, prec_ap_true_active, color="black", label="AP, NA True: "+str(np.round(prarea_ap_true_active,6)))
ax3.plot(rec_na_true_active, prec_na_true_active, color="blue", label="NA True: "+str(np.round(prarea_na_true_active,6)))
ax3.plot(rec_nona_true_active, prec_nona_true_active, color="green", label="No NA True: "+str(np.round(prarea_nona_true_active,6)))
ax3.plot(rec_comb_pred_active, prec_comb_pred_active, color="red", label="Comb Pred: "+str(np.round(prarea_comb_pred_active,6)))
ax3.plot(rec_whole_pred_active, prec_whole_pred_active, color="purple", label="Whole Pred: "+str(np.round(prarea_whole_pred_active,6)))
ax3.set_xlabel("Recall")
ax3.set_ylabel("Precision")
ax3.spines.right.set_visible(False)
ax3.spines.top.set_visible(False)
ax3.legend()

ax4.scatter(np.arange(0,500,2), spk_asoma[2350:2600]*2-0.8, marker="|", s=250, color="black", label="Measured APs")
ax4.plot(np.arange(0,500,2), ap_true_asoma[2350:2600], linewidth=1, color="black", alpha=1, label="AP, No NA True")
ax4.plot(np.arange(0,500,2), na_true_asoma[2350:2600], linewidth=1, color="blue", alpha=0.75, label="NA True")
ax4.plot(np.arange(0,500,2), nona_true_asoma[2350:2600], linewidth=1, color="green", alpha=0.75, label="No NA True")
ax4.plot(np.arange(0,500,2), comb_pred_asoma[2350:2600], linewidth=1, color="red", alpha=0.75, label="Comb Pred")
ax4.plot(np.arange(0,500,2), whole_pred_asoma[2350:2600], linewidth=1, color="purple", alpha=0.75, label="Whole Pred")
ax4.legend()
ax4.set_ylim(0,1.5)
ax4.set_xlabel("Time (ms)")
ax4.set_ylabel("AP Probability")
ax4.set_yticks([0,0.2,0.4,0.6,0.8,1])
ax4.spines.right.set_visible(False)
ax4.spines.top.set_visible(False)
ax4.spines.left.set_bounds((0, 1))

bp1 = ax5.boxplot(prarea_ap_true_asoma_array, positions=[1], widths = 0.6, patch_artist=True)
bp2 = ax5.boxplot(prarea_na_true_asoma_array, positions=[2], widths = 0.6, patch_artist=True)
bp3 = ax5.boxplot(prarea_nona_true_asoma_array, positions=[3], widths = 0.6, patch_artist=True)
bp4 = ax5.boxplot(prarea_comb_pred_asoma_array, positions=[4], widths = 0.6, patch_artist=True)
bp5 = ax5.boxplot(prarea_whole_pred_asoma_array, positions=[5], widths = 0.6, patch_artist=True)
bp1['boxes'][0].set_facecolor("black")
bp2['boxes'][0].set_facecolor("blue")
bp3['boxes'][0].set_facecolor("green")
bp4['boxes'][0].set_facecolor("red")
bp5['boxes'][0].set_facecolor("purple")
bp1['boxes'][0].set_alpha(0.5)
bp2['boxes'][0].set_alpha(0.5)
bp3['boxes'][0].set_alpha(0.5)
bp4['boxes'][0].set_alpha(0.5)
bp5['boxes'][0].set_alpha(0.5)
bp1['medians'][0].set_color("black")
bp2['medians'][0].set_color("black")
bp3['medians'][0].set_color("black")
bp4['medians'][0].set_color("black")
bp5['medians'][0].set_color("black")
ax5.set_xticklabels(["AP, No NA True","NA True", "No NA True","Comb Pred","Whole Pred"])
ax5.set_ylabel("Precision-Recall AUC")
ax5.spines.right.set_visible(False)
ax5.spines.top.set_visible(False)

ax6.plot(rec_ap_true_asoma, prec_ap_true_asoma, color="black", label="AP, No NA True: "+str(np.round(prarea_ap_true_asoma,6)))
ax6.plot(rec_na_true_asoma, prec_na_true_asoma, color="blue", label="NA True: "+str(np.round(prarea_na_true_asoma,6)))
ax6.plot(rec_nona_true_asoma, prec_nona_true_asoma, color="green", label="No NA True: "+str(np.round(prarea_nona_true_asoma,6)))
ax6.plot(rec_comb_pred_asoma, prec_comb_pred_asoma, color="red", label="Comb Pred: "+str(np.round(prarea_comb_pred_asoma,6)))
ax6.plot(rec_whole_pred_asoma, prec_whole_pred_asoma, color="purple", label="Whole Pred: "+str(np.round(prarea_whole_pred_asoma,6)))
ax6.set_xlabel("Recall")
ax6.set_ylabel("Precision")
ax6.spines.right.set_visible(False)
ax6.spines.top.set_visible(False)
ax6.legend()

fig.tight_layout()
#plt.show()
#fig.savefig("/home/sklee/dendrite/fig8/fig8_raw.pdf", bbox_inches="tight", transparent=True)
```
